# 📊 Day 21: Exploratory Data Analysis (EDA) & Statistical Insights

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student_data.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df["Gender"].value_counts()
```

```python
df.groupby("Gender")["Marks"].mean()
```

```python
sns.histplot(df["Marks"], kde=True)
plt.title("Marks Distribution")
plt.show()
```

```python
sns.scatterplot(x="Study_Hours", y="Marks", data=df)
plt.title("Study Hours vs Marks")
plt.show()
```

```python
sns.scatterplot(x="Attendance", y="Marks", data=df)
plt.title("Attendance vs Marks")
plt.show()
```

```python
sns.boxplot(x="Gender", y="Marks", data=df)
plt.title("Marks by Gender")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
print("Mean Marks:", df["Marks"].mean())
print("Median Marks:", df["Marks"].median())
print("Standard Deviation:", df["Marks"].std())
```

📌 Key Insights

1️⃣ Students with higher study hours tend to score higher marks, indicating a strong positive relationship.

2️⃣ Attendance also positively impacts performance, suggesting consistency improves results.

3️⃣ The dataset shows moderate variability, meaning most students perform within a stable range.

4️⃣ Female students show slightly higher average marks compared to male students.

5️⃣ A few students with low study hours and attendance fall into the low-performance category, indicating key improvement areas.
